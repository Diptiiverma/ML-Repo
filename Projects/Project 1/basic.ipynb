```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error

data = pd.read_excel("students.xlsx")

X = data[['Hours']] #2D input
y = data['Score']   #Target column

model = LinearRegression()
model.fit(X,y)

predicated_score = model.predict(X)

#evaluate
mae = mean_absolute_error(y , predicated_score)
mse = mean_squared_error(y , predicated_score)
rmse = np.sqrt(mse)

print("Mean Absolute ERROR (MAE) : " ,mae)
print("Mean Squared ERROR (MSE) : " ,mse)
print("Root Mean Squared ERROR (RMSE) : " ,rmse)

new_prediction = float(input("Enter a hour = "))
new_pred = model.predict([[new_prediction]])
print(f"Prediction for {new_prediction} is score = {new_pred}")

```
