```python
#import packages and libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score

#dataset load
data = pd.read_csv("student_study_dataset.csv")

#input and output
X = data[['study_hours_per_week']]
y = data['Final_score']

#train model
model = LinearRegression()
model.fit(X,y)
predicated_scores = model.predict(X)

#Valid regression matrix

#evaluations
mae = mean_absolute_error(y , predicated_scores)
mse = mean_squared_error(y , predicated_scores)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicated_scores)

#result

"""
3.124627 = round(3.124627 , 2) = 3.12
"""
print("Mean Absolute ERROR (MAE) : " ,round(mae , 2) )
print("Mean Squared ERROR (MSE) : " ,round(mse , 2) )
print("Root Mean Squared ERROR (RMSE) : " ,round(rmse , 2) )
print("R^2 Score (Model Accuracy) : " , round(r2 , 4) )


#histogram
plt.figure(figsize= (10,6))
plt.hist(data["Final_score"] , bins=30 , color='skyblue' , edgecolor = 'black')
plt.title("Distribution of Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()

#scatter + regression line
plt.figure(figsize= (10,6))
plt.scatter(X , y , color= 'blue' , label= 'Actual Scores')
plt.plot(X , predicated_scores , color = 'red' , label= 'Predicted Scores (Regression Line)')
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours per week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

#prediction
new_hours = 9
predicted_new_score = model.predict([[new_hours]])
print(f"Predicted Final Score for {new_hours} Hours is {predicted_new_score} .")

```
